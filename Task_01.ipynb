```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
 # Adjust the path if needed
df = pd.read_csv("svm_data.csv")


# Display basic info (optional)
print(df.info())
print(df.head())

# Set style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram for 'x1'
sns.histplot(df['x1'], bins=10, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of x1')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('Frequency')

# Bar chart for 'y'

sns.countplot(x='y', hue='y', data=df, ax=axes[1], palette='pastel', legend=False)
axes[1].set_title('Distribution of y (Categories)')
axes[1].set_xlabel('Category (y)')
axes[1].set_ylabel('Count')


# Adjust layout
plt.tight_layout()
plt.show()
```
